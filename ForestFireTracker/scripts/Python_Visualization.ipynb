```python
#Visualization 1 for question 4 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# data = pd.read_csv("populations_endangered.csv", error_bad_lines=False)
data = pd.read_csv("populations_joined.csv", error_bad_lines=False)

data.rename(columns={'species':'Species',
                          'fname':'Forest_Name',
                          'density':'Density',
                    'headcount': 'Headcount', 'endangered':'Endangered'}, 
                 inplace=True)
```

```python
data
```

```python
data.drop(7, inplace=True)
data = data.reset_index(drop=True)
```

```python
data
```

```python
data_endangered = data[data.Endangered.eq('t')] #Create frame for endangered species
```

```python
data_endangered
# groupedvalues = data_endangered.groupby('Species').sum()
# groupedvalues
```

```python
import seaborn as sns
sns.set(style="whitegrid")

g = sns.barplot(x=data_endangered.Species, y=data_endangered.Headcount,
                 hue=data_endangered.Forest_Name)

g.set_title('Endangered Species Headcount by Forest Name')

ax = g


for p in ax.patches:
    height = p.get_height()

    ax.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:1.1f}'.format(height),
            ha="center") 
    
plt.setp(ax.get_xticklabels(), rotation=30, horizontalalignment='right')
```
